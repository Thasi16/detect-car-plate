```python
from ultralytics import YOLO
import os
import cv2

# Load model đã train
model = YOLO(r"C:\Users\Admin\Desktop\dpl\runs\detect\train5\weights\best.pt")

# thư mục chứa ảnh
input_folder = r"C:\Users\Admin\Desktop\dpl\cropped_vehicles"

# thư mục lưu biển số đã crop
output_folder = r"C:\Users\Admin\Desktop\plates"
os.makedirs(output_folder, exist_ok=True)

image_ext = (".jpg", ".jpeg", ".png")

for file in os.listdir(input_folder):
    if file.lower().endswith(image_ext):

        img_path = os.path.join(input_folder, file)
        img = cv2.imread(img_path)

        results = model(img)

        boxes = results[0].boxes.xyxy.cpu().numpy()

        for i, box in enumerate(boxes):
            x1, y1, x2, y2 = map(int, box)

            crop = img[y1:y2, x1:x2]

            save_name = f"{os.path.splitext(file)[0]}_plate_{i}.jpg"
            save_path = os.path.join(output_folder, save_name)

            cv2.imwrite(save_path, crop)

        print(f"Processed {file}")

print("Crop hoàn tất!")
```
